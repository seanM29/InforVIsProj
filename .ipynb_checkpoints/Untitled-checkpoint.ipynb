```python
import pandas as pd
import numpy
import datetime
from os import listdir
from os.path import isfile, join
import os, errno
```

```python

def preprocessData(filename,Savedir):
    try:
        os.makedirs(Savedir)
    except OSError as e:
        if e.errno != errno.EEXIST:
            raise
    
    fielpath = "./data_raw/"+filename
    df=pd.read_csv(fielpath)
    tf = df.values
    list_hour=[]
    for row in tf:
        d=datetime.datetime.fromtimestamp(int(row[4]/1000))
        list_hour.append(d.hour)
    se = pd.Series(list_hour)
    df['hour']=se.values
    for i in range(0,24): 
        savepath_csv=Savedir+str(i)+".csv"
        savepath_json=Savedir+str(i)+".json"
        df.loc[df['hour'] == i].to_csv(savepath_csv,index=None)
        df.loc[df['hour'] == i].to_json(savepath_json,orient = "records")
    savepath_all_csv=Savedir+"/all.csv"
    df.to_csv(savepath_all_csv)
    print "Finish "+filename
```

```python
mypath="./data_raw"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))] ##get all file
for i in onlyfiles:
    FileName=i
    ##print FileName
    DirName ="./Data/" +FileName.split('.')[0]+"/"
    ##print DirName
    preprocessData(FileName,DirName)
```

```python
##get type

def getType(Content):
    result=-1;
    if(Content.find("白领")>=0):
        result=0;
    elif (Content.find("上门服务")>=0):
        result=0;
    
    elif (Content.find("票")>=0):
        result=1;
   
        
    elif (Content.find("点低")>=0):
        result=1;
    elif(Content.find("现房")>=0):
        result=2;
    elif(Content.find("万")>=0):
        result=2;
    elif(Content.find("合同")>=0):
        result=2;
    elif(Content.find("送")>=0):
        result=2;
    elif(Content.find("驾校")>=0):
        result=2;
    elif(Content.find("特价")>=0):
        result=2;
    elif(Content.find("低折")>=0):
        result=2;
    elif(Content.find("免费")>=0):
        result=2;
      
    elif(Content.find("信用卡")>=0):
        result=3;    
    elif(Content.find("行")>=0):
        result=3;
    elif(Content.find("抄盘手")>=0):
        result=3;
    elif(Content.find("办证")>=0):
        result=4;
    elif(Content.find("博彩")>=0):
        result=5;
    elif(Content.find("赌场")>=0):
        result=5;
    elif(Content.find("驾驶证")>=0):
        result=6;
    elif(Content.find("贷")>=0):
        result=7;
    else:
        result=8;
    return result
    
def get_type_of_day(filename):
    fileindex=str(filename).split('.')[0]
    readfile="./Data/"+str(fileindex)+"/all.csv"
    dataframe=pd.read_csv(readfile)
    ll=dataframe.values.tolist()
    a = numpy.zeros(shape=(24,9))
    i=0
    while i<len(ll):
        ContentType=getType(ll[i][2])
        ##print str(i)+"#"+str(ContentType)+"#"+str(ll[i][8])
        a[ll[i][8]][ContentType]+=1
        i=i+1
    d=pd.DataFrame(data=a)
    save_json="./Data/Chart/Type/"+str(fileindex)+".json"
    d.to_json(save_json,orient = "records")    
   ## return d
```

```python
mypath="./data_raw"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))] ##get all file
for i in onlyfiles:
    get_type_of_day(i)
    print "finsh "+i
```
